###**Extracting Wikipedia Page Name from the Page Column**
---


```python
def extract_name(page):
    pattern = r'(.{0,})_(.{2}).wikipedia.org_'
    result = re.findall(pattern, page)
    if len(result) == 1:
        return result[0][0]
    else:
        return 'unknown'
```

```python
df['name'] = df['Page'].apply(extract_name)
```

#Explanation:
- **Def Function takes on epage string at a Time**
- **Regex pattern:**
  - **(.{0,}) - Capture Everything Before the Language Code**
  - **_ : Underscore separator**
  - **(.{2}) : Capture 2-Letter language code (en, ja, ru)**
  - **.wikipedia.org_ : Fixed Wikipedia text**
- **It Captures : (Page_Name) _ (Language).wikipedia.org_**
---
>**EXTRACT MATCHING TEXT:**
- **result = re.findall(pattern, page)**
  - **Searches for the Pattern**
  - **Returns a list of matches**
---
>**SAFETY CHECK**
- **If Match is Found : Return Page name**
- **If Not : Return Unknown : To avoid Errors**
---
>**APPLY FUNCTION**
- **Applies Function to every row**
- **Creates a New column called name**
---

---
###**Extract Language Code from Wikipedia Page Name**
---

```python
def extract_lang(page):
    pattern = r'(.{0,})_(.{2}).wikipedia.org_'
    result = re.findall(pattern, page)
    if len(result) == 1:
        return result[0][1]
    else:
        return 'un'

df['language'] = df['Page'].apply(extract_lang)
df['language'].unique()
```

---
###**Mapping Language Codes to Full Language Names**
---


```python
lang_name_mapping = {
    'zh': 'Chinese',
    'fr': 'French',
    'en': 'English',
    'un': 'Unknown',
    'ru': 'Russian',
    'de': 'German',
    'ja': 'Japanese',
    'es': 'Spanish'
}

df['language'] = df['language'].map(lang_name_mapping)
```

#Explanation:
- **Converts short language codes like en, ja, zh**
- **Into readable names like English, Japanese, Chinese**
- **Makes Plot and insights eady to understand**

###**Stationarity Check Using Augmented Dickey-Fuller (ADF) Test for Each Language**
---
>**Before Applying ARIMAX or SARIMAX The Time series must be stationary**
- **This Code checks whether page views for each Language are stationary or Not**
---
###**Why stationarity is Important**
---
- **ARIMA Models assume stable mean and variance**
- **Non-stationary data leads to poor forecasts**
- **Making data stationary improves Prediction accuracy**
---

---
###**Step 1: Define ADF Test Function**
---

```python
def adfuller_test(time_series):
    p_value = sm.tsa.stattools.adfuller(time_series)[1]
    if(p_value < 0.05):
        print('The time series is stationary')
    else:
        print('The time series is not stationary')
```

#Explanation:
- **Runs the ADF Test on the given time series**
- **Extracts the P-value**
>**Decision Based on p-value:**
- **p-value < 0.05 : data is stationary**
- **p-value >= 0.05 : data is not stationary**

```python
for lang in df_agg.columns:
    print(lang)
    adfuller_test(df_agg[lang])
    print()
```

#Observations:
- **Most languages (Chinese, English, French, German, Japanese) have non-stationary traffic that changes over time.**
- **Russian, Spanish, and Unknown languages have stationary traffic, staying more consistent around a stable average.**

###**ACF and PACF Plots for ARIMA Model Identification**
---
###**Autocorrelation Function (ACF)**
---
>**The ACF measures the correlation between a time series and its own past values at different time lags.**
- **It shows how strongly the current value depends on previous values**
- **Used to identify the Moving Average (MA – q) order**
- **A sharp cutoff after lag q suggests an MA(q) model**
- **Shows how the time series is correlated with its past values**
- **Checks correlation at lag 1, lag 2, lag 3, …**
- **Used to decide MA (q) order**
- **If ACF cuts off after q lags → MA(q)**
---
###**Partial Autocorrelation Function (PACF)**
---
>**The PACF measures the direct correlation between a time series and its past values at a given lag, after removing the effect of all intermediate lags.**
- **It shows the pure impact of a specific lag**
- **Used to identify the Autoregressive (AR – p) order**
- **A sharp cutoff after lag p suggests an AR(p) model**
- **Shows direct correlation with a lag**
- **Removes the effect of intermediate lags**
- **Used to decide AR (p) order**
- **If PACF cuts off after p lags → AR(p)**
---

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# First-order differenced (stationary) series
ts_diff = ts_english.diff(1).dropna()

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 4))

# PACF plot → AR order (p)
plot_pacf(
    ts_diff,
    ax=axs[0],
    lags=30,
    color='#1b5e20'
)
axs[0].set_title('PACF Plot (For AR Order p)', fontweight='bold')
axs[0].set_xlabel('Lag')
axs[0].set_ylabel('Partial Correlation')

# ACF plot → MA order (q)
plot_acf(
    ts_diff,
    ax=axs[1],
    lags=30,
    color='#0d47a1'
)
axs[1].set_title('ACF Plot (For MA Order q)', fontweight='bold')
axs[1].set_xlabel('Lag')
axs[1].set_ylabel('Correlation')

plt.tight_layout()
plt.show()
```

#Observations:
- **Mostly Random: Both PACF and ACF show most bars inside the shaded area, meaning the data is mostly white noise.**
- **Weak Memory: PACF has almost no spikes, so today’s value isn’t strongly tied to past values.**
- **Tiny Weekly Pattern: Small spikes at Lags 7, 14, and 21 suggest a slight weekly cycle.**
- **Short Relationships: Even the weekly pattern is weak, so it won’t strongly help predictions.**
- **Modeling Tip: Start simple (like ARIMA(0,0,0)) since the series is mostly unpredictable.**

###**Model Performance Evaluation Function**
---
###**What this Code is Doing**
---
>**This Function is used to check how good a Forecasting Model is by Comparing**
- **Actual values (Real Page views)**
- **Predicted values (Model's Forecast)**
- **It calculates three popular error metrics**
----

```python
# Creating a function to print values of all these metrics.
def performance(actual, predicted, print_metrics=True):
    MAE = round(mae(actual, predicted), 3)
    RMSE = round(mse(actual, predicted)**0.5, 3)
    MAPE = round(mape(actual, predicted), 3)
    if(print_metrics==True):
        print('MAE :', MAE)
        print('RMSE :', RMSE)
        print('MAPE:', MAPE)
    return MAE, RMSE, MAPE
```

#Explanation:
- **Mean Absolute Error:**
  - **Average of absolute errors, tells how far predictions are from actual values (On average)**
- **Root Mean Squared Error:**
  - **Penalizes large errors more heavily**
  - **Square root makes it interpretable in original units**
- **Mean absolute percentage error:**
  - **Shows error in percentage terms**
  - **Easy for business users to understand**
---
###**Why This Function is Important**
---
- **Standard way to evaluate forecasting models**
- **Helps compare ARIMA vs other models**
- **Useful for business decisions (ad planning, traffic forecasting)**
---

  - **No MA Terms**

  - **MAE : Average absolute Error**
  - **RMSE : Penalizes Large Errors**

###**BUSINESS MEANING**
---
>**This model helps predict future English Wikipedia traffic, which can be used to:**
- **Plan ad placements**
- **Allocate server resources**
- **Schedule marketing campaigns**
- **Estimate future user engagement**

```python
# Forecast horizon
n_forecast = 60

# Train ARIMA model (d=1 since series is stationary after differencing)
model = ARIMA(TS[:-n_forecast], order=(0, 1, 0))
model = model.fit()

# Generate forecasts
predicted = model.forecast(steps=n_forecast, alpha=0.05)

# Plot actual vs forecast
plt.figure(figsize=(11,4))

TS.plot(
    label='Actual Page Views',
    color='#1f77b4',
    linewidth=2
)

predicted.plot(
    label='Forecasted Page Views',
    linestyle='dashed',
    marker='o',
    color='#ff7f0e'
)

plt.title(
    'ARIMA (0,1,0) Forecast vs Actual\nEnglish Wikipedia Page Views',
    fontsize=13,
    fontweight='bold'
)

plt.xlabel('Date', fontweight='bold')
plt.ylabel('Number of Page Views', fontweight='bold')

plt.legend(loc="upper right")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Model performance
(_, _, _) = performance(
    TS.values[-n_forecast:],
    predicted.values,
    print_metrics=True
)
```

#Observations:
- **Random Walk Model: The ARIMA(0,1,0) forecast basically predicts tomorrow will be the same as today, so the forecast line is flat.**
- **Volatile Data: Actual views (blue line) have many spikes and dips, especially around July (4,000 → 8,000).**
- **Forecast Limitations: The flat forecast misses all the big swings and only captures the average trend.**
- **Performance Metrics: On average, the forecast is off by ~478 views (MAE), with bigger misses raising RMSE to ~673, and about 8.6% error (MAPE).**
- **Verdict: The model gives a stable baseline but cannot predict spikes; better results need seasonality or event-based adjustments.**

- **Captures:**
  - **Trend (d = 1)**
  - **Weekly seasonality (S = 7)**
  - **Campaign Impact (Exogenous variable)**
- **Forecasts the next 60 days of page views**
---
###**Why SARIMAX is better than ARIMA here**
---
| Model       | What it Uses                           |
| ----------- | -------------------------------------- |
| ARIMA       | Only past page views                   |
| SARIMA      | Past views + seasonality               |
| **SARIMAX** | Past views + seasonality + campaigns ✅ |
---
###**Business Insight**
---
- **During Campaign days page views spike**
- **SARIMAX Learns this effect and predicts more accurately**
- **Helps AdEase**
  - **Place Premium ads on high-impact days**
  - **Avoid overspending on low-impact days**
---

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Exogenous variable (campaign flag: 0 = no campaign, 1 = campaign)
exog = exog_en['Exog'].to_numpy()

# Model parameters
p, d, q = 1, 1, 1
P, D, Q, s = 1, 1, 1, 7   # weekly seasonality
n_forecast = 60

# Build SARIMAX model
model = SARIMAX(
    TS[:-n_forecast],
    order=(p, d, q),
    seasonal_order=(P, D, Q, s),
    exog=exog[:-n_forecast],
    initialization='approximate_diffuse'
)

# Fit model
model_fit = model.fit()

# Forecast future values
model_forecast = model_fit.forecast(
    steps=n_forecast,
    dynamic=True,
    exog=pd.DataFrame(exog[-n_forecast:])
)

# Plot Actual vs Forecast
plt.figure(figsize=(10,5))

TS[-120:].plot(
    label='Actual Page Views',
    color='#1f77b4',
    linewidth=2
)

model_forecast[-120:].plot(
    label='Forecasted Page Views',
    color='red',
    linestyle='dashed',
    marker='o',
    markerfacecolor='green'
)

plt.title(
    f'SARIMAX Model ({p},{d},{q}) × ({P},{D},{Q},{s})\nActual vs Forecast (English Pages)',
    fontsize=14,
    fontweight='bold'
)

plt.xlabel('Date', fontweight='bold')
plt.ylabel('Number of Page Views', fontweight='bold')

plt.legend(loc='upper right')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Model performance
(_, _, _) = performance(
    TS.values[-n_forecast:],
    model_forecast.values,
    print_metrics=True
)
```

###**SARIMAX Hyperparameter Search Function (Model Tuning)**
---
>**WHAT IS THE USE OF THIS CODE**
- **The Function is used to Find the best SARIMAX Model parameter by trying many combinations of:**
  - **ARIMA Parameters - (p, d, q)**
  - **Seasonal Parameters - (P, D, Q, S)**
  - **External Factors - campaign effect (exog)**
- **It helps us choose the model that gives the lowest forecasting error**
---
>**WHAT THIS FUNCTION DOES**
- **Takes the English time series and campaign data**
- **Tries all combinations of SARIMAX parameters you give**
- **For Each combination:**
  - **Trains a SARIMAX Model**
  - **Forecasts the next n_forecast days**
  - **Calculate MAE, RMSE, and MAPE**
- **Stores the results in a DataFrame**
- **Returns a table to compare all models**
---
>**WHY THIS IS IMPORTANT FOR THE BUSINESS**
- **Different languages & regions have differently**
- **One SARIMAX Model wont fit all**
- **This Function helps:**
  - **Pick the best Model**
  - **Improve accuracy (Target MAPE : 4-8%)**
  - **Make better ad-placement decisions**
---


```python
def SARIMAX_search(TS, forecast, p_list, d_list, q_list, P_list, D_list, Q_list, s_list, exog=[]):
    counter = 0
    #perf_df = pd.DataFrame(columns=['serial', 'pdq', 'PDQs', 'mape', 'rmse', 'aic', 'bic'])
    perf_df = pd.DataFrame(columns=['serial', 'pdq', 'PDQs', 'mape', 'rmse'])

    for p in p_list:
        for d in d_list:
            for q in q_list:
                for P in P_list:
                    for D in D_list:
                        for Q in Q_list:
                            for s in s_list:
                                try:
                                    model = SARIMAX(TS[:-n_forecast], order =(p,d,q), seasonal_order=(P, D, Q, s), exog = exog[:-n_forecast], initialization='approximate_diffuse')
                                    model_fit = model.fit()
                                    model_forecast = model_fit.forecast(n_forecast, dynamic = True, exog = pd.DataFrame(exog[-n_forecast:]))
                                    MAE, RMSE, MAPE = performance(TS.values[-n_forecast:], model_forecast.values, print_metrics=False)
                                    counter += 1
                                    #list_row = [counter, (p,d,q), (P,D,Q,s), MAPE, RMSE, model_fit.aic, model_fit.bic]
                                    list_row = [counter, (p,d,q), (P,D,Q,s), MAPE, RMSE]
                                    perf_df.loc[len(perf_df)] = list_row
                                    print(f'Combination {counter} out of {(len(p_list)*len(d_list)*len(q_list)*len(P_list)*len(D_list)*len(Q_list)*len(s_list))}')
                                except:
                                    continue
    return perf_df
```

---
###**Running SARIMAX Grid Search for Optimal Model Selection**
---
>**This Code block is used to execute a controlled grid search over different SARIMAX Parameter Combinations to identify the best-performing Model Based on error Metrics like MAPE and RMSE**
---

```python
if 0:
    TS = ts_english.copy(deep=True)
    n_forecast = 60
    p_list = [0,1]
    d_list = [1]
    q_list = [0,1]
    P_list = [2,3]
    D_list = [1]
    Q_list = [2,3]
    s_list = [7]
    exog = exog_en['Exog'].to_numpy()
    perf_df = SARIMAX_search(TS, n_forecast, p_list, d_list, q_list, P_list, D_list, Q_list, s_list, exog)
    perf_df.sort_values(['mape', 'rmse'])
```
